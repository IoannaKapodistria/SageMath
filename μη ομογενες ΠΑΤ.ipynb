- (α) Χρησιμοποιώντας το Sage βρίσκουμε την λύση του μη ομογενούς συστήματος <br><br>
$$ y_1'(t ) = \,\,\,\,y_2(t)\,, \phantom{+ 2\,\cos t}\\ 
   y_2'(t ) = -y_1(t) + 2\,\cos t\,, $$ <br>
με αρχικές συνθήκες $y_1(0)=4$, $y_2(0)=0$.<br><br>
- (β) Σχεδιάζουμε τις λύσεις $y_1(t)$, $y_2(t)$, ως συναρτήσεις το χρόνου $t$, στο ίδιο γραφικό με διαφορετικό χρώμα η καθεμιά. Τι είδους ταλαντώσεις εκτελούν οι λύσεις;<br><br>
- (γ) Σχεδιάζουμε την λύση στο επίπεδο $(y_1,y_2)$ παραμετρικά ως προς $t$.

## Άσκηση 2 
Θεωρούμε τον πίνακα Α
$$ Α = \left( \begin{array}{r r r r} 
0  & 1  & 0 & 0 \\
-2 & -1 & 1 & 0 \\
0  &  0 & 0 & 1 \\
1  & 0  &-2 & -1 \end{array}\right) \,.
$$
- α) Βρίσκουμε τις ιδιοτιμές και τα αντίστοιχα ιδιοδιανύσματα του A, χρησιμοποιώντας τις αντίστοιχες εντολές του Sage.
- β) Λύνουμε το ΠΑΤ για το σύστημα ΣΔΕ 
$$ \mathbf{y}' = A\, \mathbf{y}\,,\qquad \mathbf{y}(0)= (2,1,-2,1)\,,$$
χρησιμοποιώντας την εντολή `desolve_system` . 
- γ) Σχεδιάζουμε την λύση του β) υποερωτήματος στο επίπεδο $(y_1,y_2)$. Πρόκειται για σπείρα που συγκλίνει στο σημείο $(0,0)$. Για ποιόν λόγο συμβαίνει αυτό; Έχει να κάνει με το πρόσημο του πραγματικού μέρους των ιδιοτιμών;

## Άσκηση 1

```python
var('t'); 
y1 = function('y1')(t); 
y2 = function('y2')(t);
y = vector([y1,y2])
A = matrix([[0,1],[-1,0]])
c = vector([0, 2*cos(t)])
system = [diff(y[i], t) - (A * y)[i] - c[i]  for i in range(2)]
sol = desolve_system(system, [y1, y2], ivar=t, ics=[0,4,0] ) ; print sol
```

```python
p1 = plot(sol[0].rhs(),(t,0,50),color='red'); 
p2 = plot(sol[1].rhs(),(t,0,50),color='blue'); 
(p1+p2).show(figsize=5)
```

```python
y1y2 = parametric_plot([sol[0].rhs(),sol[1].rhs()],(t,0,50)); y1y2.show(figsize=5)
```
